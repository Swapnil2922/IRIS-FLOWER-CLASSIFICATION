# IRIS FLOWER CLASSIFICATION: Data Visualization

```python
data.columns= ["SepalLengthCm","SepalWidthCm","PetalLengthCm",
               "PetalWidthCm","Species"]
plt.title("IRIS FLOWER CLASSIFICATION")
sns.histplot(x="SepalLengthCm", hue="Species", data=data)
plt.show()
```

```python
sns.countplot(x = 'Species',data=data)
plt.show()
```

```python
colors=['red','black','blue']
Species=['Iris-setosa','Iris-versicolor Species','Iris-virginica']
```

```python

sns.scatterplot(x='SepalLengthCm', y='SepalWidthCm', data=data)
plt.show()
```

```python
sns.scatterplot(x='SepalLengthCm', y='SepalWidthCm',
                hue='Species',
                data=data)
plt.show()
```

```python
sns.regplot(x='SepalLengthCm', y='SepalWidthCm', data=data)
plt.show()
```

```python
#facets

g = sns.FacetGrid(data, col="Species")
g.map(plt.hist, "SepalLengthCm")
plt.show()
```

```python
g = sns.FacetGrid(data, row="Species")
g.map(plt.hist, "SepalLengthCm")
plt.show()
```

```python
g = sns.FacetGrid(data, col="Species")
g.map(plt.scatter, "SepalLengthCm","SepalWidthCm")
plt.show()
```

```python
#jointplot

sns.jointplot(x='SepalLengthCm', y='SepalWidthCm', data=data)
plt.show()

sns.jointplot(x='SepalLengthCm', y='SepalWidthCm', data=data,kind="reg")
plt.show()
```

```python
#Pairs plot\matrix plot

sns.pairplot(data)
plt.show()
```

```python
sns.pairplot(data, hue="Species")
plt.show()
```

```python
#Heatmap

import seaborn as sns
sns.heatmap(data.corr(),annot=True)
plt.show()
```

```python
sns.heatmap(data.corr(),cmap="YlGnBu",annot=True)
plt.xticks(rotation=45)
plt.yticks(rotation=45)
plt.show()
```
